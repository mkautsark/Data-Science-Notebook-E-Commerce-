```python

def month(x):
    if x == 'JAN' or x == 'Jan':
        x = 'January'
    if x == 'FEB' or x == 'Feb':
        x = 'February'
    if x == 'MAR' or x == 'Mar':
        x = 'March'
    if x == 'APR' or x == 'Apr':
        x = 'April'
    if x == 'MAY' :
        x = 'May'
    if x == 'JUN' or x == 'jun' or x == 'Jun':
        x = 'June'
    if x == 'JUL' or x == 'jul' or x == 'Jul':
        x = 'July'
    if x == 'AUG' or x == 'Aug':
        x = 'August'
    if x == 'SEP'or x == 'Sep':
        x = 'September' 
    if x == 'OCT'or x == 'Oct':
        x = 'October'
    if x == 'NOV'or x == 'Nov':
        x = 'November'
    if x == 'DEC'or x == 'Dec':
        x = 'December'  
    return x 
```
